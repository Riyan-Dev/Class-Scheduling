```python
class Schedule:

    def __init__(self):
        self._data = data
        self. _classes = []
        self._numbOfConflicts = 0
        self._fitness = -1
        self._classNumb = 0
        self._isFitnessChanged = True
        

    def get_classes (self):
        self. isFitnessChanged = True
        return self._classes
    
    def get_numbOfConflicts (self): return self._numbOfConflicts 
    def get_fitness(self):
        if (self._isFitnessChanged == True):
            self._fitness = self.calculate_fitness ()
            self. isFitnessChanged = False
        return self._fitness
    
    def get_binaryEncodings(self):
        self._binaryEncodings = []
        for i in range(0, len(self._classes)):
            self._binaryEncodings.append(self._classes[i].get_binaryCode())
        
        return self._binaryEncodings
    
    def initialize(self):
        
        sections = self._data.get_sections()
        for z in range(0, len(sections)):
            depts = self._data.get_depts()

            for i in range(0, len(depts)):
                courses = depts[i].get_courses()
                for j in range(0, len (courses)):
                    self._classNumb += 1
                    if 'L' in courses[j].get_number():
                        newClass = Class(self._classNumb, depts[i], courses[j], sections[z], courses[j].get_instructors()[rnd.randrange(0, len(courses[j].get_instructors ()))],  data.get_labTimings()[rnd.randrange(0,len(data.get_labTimings()))], data.get_rooms ()[rnd.randrange(0, len(data.get_rooms ()))])

                        
                    else:
                        newClass = Class(self._classNumb, depts[i], courses[j], sections[z], courses[j].get_instructors()[rnd.randrange(0, len(courses[j].get_instructors ()))],  data.get_meetingTimes()[rnd.randrange(0,len(data.get_meetingTimes()))], data.get_rooms ()[rnd.randrange(0, len(data.get_rooms ()))])


                    
                    
                    self._classes.append(newClass)

        return self
    
    def binaryDecoding(self, binaryCodings):
        self._classes = []
        for i in range(0, len(binaryCodings)):
            self._classes.append(Class(i, None, None))
            self._classes[i].binaryDecoding(binaryCodings[i]) 
        
        # return self
    def binaryDecodingAfterMutation(self):
        for i in range(0, len(self._classes)):
            self._classes[i].binaryDecoding(self._binaryEncodings[i])
    
    def is_consecutive_classes(self, mt1, mt2):
        # Extract day codes from meeting times
        day_code_1 = mt1.get_days()
        day_code_2 = mt2.get_days()

        # Check if both meeting times are on the same day
        if day_code_1 != day_code_2:
            return False
        # Extract start and end times from meeting times
        start_time_1 = mt1.get_time().split(" - ")[0]
        end_time_1 = mt1.get_time().split(" - ")[1]

        start_time_2 = mt2.get_time().split(" - ")[0]
        end_time_2 = mt2.get_time().split(" - ")[1]

        # Check if the end time of mt1 is the same as the start time of mt2
        if end_time_1 == start_time_2:
            return True

        # Check if the end time of mt1 is before the start time of mt2 with a buffer
        buffer_minutes = 10  # Adjust as needed
        end_time_1_minutes = int(end_time_1.split(":")[0]) * 60 + int(end_time_1.split(":")[1])
        start_time_2_minutes = int(start_time_2.split(":")[0]) * 60 + int(start_time_2.split(":")[1])

        if end_time_1_minutes + buffer_minutes >= start_time_2_minutes:
            return True

        return False

    def calculate_fitness(self):

        self._numbOfConflicts = 0
        bonus = 0.0

        classes = self.get_classes ()

        # Check for 3 Instructors per class
        instructors = self._data.get_instructors()
        for i in range (0, len(instructors)):
            instructorCount = 0
            for j in range(0, len(classes)):
                if classes[j].get_instructor() == instructors[i]:
                    instructorCount += 1

            if (instructorCount > 3):
                self._numbOfConflicts += 1
            


        for i in range(0, len (classes)):

            # Check for checking section strength and hall capacity 
            if (classes [i].get_room().get_seatingCapacity() < classes[i].get_section().get_strength ()): 
                self._numbOfConflicts += 1
                # print("Capcity")
            
            for j in range(0, len(classes)):

                if (j >= i):
                    #  Check for same time of lab and courses
                    if ((classes [i].get_meetingTime() == classes [j].get_meetingTime() or 
                        classes[j].get_meetingTime().get_id() in self._data.get_overlaps(classes[i].get_meetingTime().get_id())) and
                        classes[i].get_id() != classes [j].get_id()):

                        # if same timing check if same room 
                        if (classes[i].get_room() == classes [j].get_room()): 
                            self._numbOfConflicts += 1 
                            # print(f"Same Room {classes[i].get_room().get_number()}")

                        # if same timing check if same Instructor 
                        if (classes [i].get_instructor() == classes[j].get_instructor ()): 
                            self._numbOfConflicts += 1 
                            # print(f"same instruct {classes [i].get_instructor()}" )

                        # If same timing check same section 
                        if (classes[i].get_section().get_section() == classes[j].get_section().get_section()):
                            self._numbOfConflicts += 1
                            # print(f" Same Sec {classes[i].get_section().get_section()}")
                            
                
                    # //Soft Constraints
                    # CHecking if labs are in the evening and all classes are in the morning 
                    if 'L' in classes[i].get_course().get_number():
                        if classes[i].get_meetingTime().get_id() in ['MT15', 'MT18', 'MT21', 'MT24', 'MT31']:
                            bonus += 0.003
                    else:
                        if classes[i].get_meetingTime().get_id() in ['MT1', 'MT2', 'MT3', 'MT4', 'MT7', 'MT8', 'MT9', 'MT10', 'MT25', 'MT26', 'MT27']:
                            bonus += 0.001
                    
                    # check for consecutive classes
                    if ((classes [i].get_instructor() == classes[j].get_instructor ()) and 
                            classes[i].get_course().get_number() == classes[j].get_course().get_number()):

                        if (self.is_consecutive_classes(classes[i].get_meetingTime(), classes[j].get_meetingTime())):
                            bonus += 0.003
                        else:
                            bonus -= 0.005
                    

                    # soft constraint for students adjacent classes having classes on same floor 
                    if (classes[i].get_section().get_section() == classes[j].get_section().get_section() and
                        self.is_consecutive_classes(classes[i].get_meetingTime(), classes[j].get_meetingTime()) and 
                        classes[i].get_room().get_floor() == classes[j].get_room().get_floor()):
                        bonus += 0.004
                    

                    # soft constraint for teachers adjacent classes having classes on same floor 
                    if (classes[i].get_instructor() == classes[j].get_instructor() and
                        self.is_consecutive_classes(classes[i].get_meetingTime(), classes[j].get_meetingTime()) and 
                        classes[i].get_room().get_floor() == classes[j].get_room().get_floor()):
                        bonus += 0.004
                    

                        
                    #     if classes[j].get_meetingTime().get_id() in self._data.get_overlaps(classes[i].get_meetingTime().get_id()):
                    #         if (classes[i].get_room() == classes [j].get_room()): 
                    #             print("lab Same Room")
                    #             self._numbOfConflicts += 1
                    #         if (classes [i].get_instructor() == classes[j].get_instructor ()): 
                    #             print("lab Same Instruct")
                    #             self._numbOfConflicts += 1
                        
                    #         if (classes[i].get_section().get_section() == classes[j].get_section().get_section()):
                    #             print("lab Same Sec")
                    #             self._numbOfConflicts += 1

        return (1 / ((1.0*self._numbOfConflicts + 1))) + bonus
    
    def __str__(self):

        returnValue = ""

        for i in range(0, len(self._classes) -1):
            returnValue += str(self._classes [i]) + ", "

        returnValue += str(self._classes [len(self._classes)-1])
        return returnValue
```

```python
class GeneticAlgo:

    def evolve(self, population): 
        # print("Hello1")
        return self._mutate_population(
            self._crossover_population(population))

    def _crossover_population (self, pop):

        crossover_pop = Population (0)

        for i in range(NUMB_OF_ELITE_SCHEDULES):
            crossover_pop.get_schedules().append(pop.get_schedules ()[i])

        i = NUMB_OF_ELITE_SCHEDULES
        
        while i < POPULATION_SIZE:
            schedule1 = self._select_tournament_population (pop) .get_schedules()[0]
            # print("Hello2")
            schedule2 = self._select_tournament_population (pop) .get_schedules ()[0]
            crossover_pop.get_schedules() .append(self._crossover_schedule(schedule1, schedule2))
            i += 1
        # print("Hello3")
        # crossover_pop.binaryDecoding()
        return crossover_pop

    def _mutate_population(self, population):
        for i in range(NUMB_OF_ELITE_SCHEDULES, POPULATION_SIZE):
            self._mutate_schedule(population.get_schedules ()[i])
        
        # population.binaryDecodingAfterMutation()
        return population
    
    def _crossover_schedule(self, schedule1, schedule2):
        # print("Hello4")
        crossoverSchedule = Schedule().initialize()

        for i in range(0, len(crossoverSchedule.get_classes())):
            if (rnd.random() > 0.5) : crossoverSchedule.get_classes()[i] = schedule1.get_classes()[i]
            else: crossoverSchedule.get_classes()[i] = schedule2.get_classes()[i]

        return crossoverSchedule
    
    def _mutate_schedule(self, mutateSchedule):

        schedule = Schedule().initialize()

        for i in range(0, len(mutateSchedule.get_classes ())) :
            if (MUTATION_RATE > rnd.random()): mutateSchedule.get_classes ()[i] = schedule.get_classes()[i]
        
        
        return mutateSchedule
    
    def _select_tournament_population(self, pop):
        # print("Hello5")
        tournament_pop = Population (0)

        i = 0

        while i < TOURNAMENT_SELECTION_SIZE:
            # print("Hello6")
            tournament_pop.get_schedules().append(pop.get_schedules ()[rnd.randrange(0, POPULATION_SIZE)])
            i += 1
            # print("Hello7")

        tournament_pop.get_schedules().sort(key=lambda x: x.get_fitness(), reverse=True)
        # print("Hello8")
        return tournament_pop
```
